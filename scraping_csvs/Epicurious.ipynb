```python
import pandas as pd
import requests
import re
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

%matplotlib inline
```

```python
url='https://www.epicurious.com/search/potatoes?page=1'
response = requests.get(url)
doc = BeautifulSoup(response.text)
regexp = re.compile(r'.*-content-card')
articles = doc.find_all(class_=regexp)
print(articles)
# for article in articles:
#     print(article.find(class_='tag').text.strip())
```

```python
print("Clearing out our dataset")
rows = []
for page_num in range(1, 11):
    
    url=f'https://www.epicurious.com/search/potatoes?page={page_num}'
    print("now scraping:",url)
    # Download the appropriate page
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    regexp = re.compile(r'.*-content-card')
    recipes = doc.find_all(class_=regexp)
#     recipes = doc.find_all(class_='recipe-content-card')
    for recipe in recipes:
        row={}
        row['tag']=recipe.find(class_='tag').text.strip()
        row['name']=recipe.find('a').text.strip()
        row['slug'] = recipe.find('a')['href']
        try:
            row['rating'] = recipe.find('span',itemprop="ratingValue").text.strip()
        except:
            pass
        row['summary'] = recipe.find(class_='dek').text.strip()
        try:
            row['make_again'] = recipe.find(class_='make-again-percentage').text
#         print(make_again)
        except:
            pass
        rows.append(row)

print("Building our dataframe")
df = pd.DataFrame(rows)
```

```python
def scrape_page(row):
    page = {}
    url = base_url + row['slug']
    print('scraping', url)
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    page['ingredients'] = doc.find(class_='ingredients').text
    page['preprations'] = doc.find(class_='preparation-steps').text.strip()
    return pd.Series(page)
    
```

```python
scraped_df = df.apply(scrape_page, axis=1)
```
