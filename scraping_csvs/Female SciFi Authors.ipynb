# Scraping Female SciFi Authors from Goodreads
Rank,Title,Author,Score,Votes,Rating

```python
import pandas as pd
import requests
import re
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

%matplotlib inline
```

```python
url='https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors'
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
books=doc.find_all('tr')
books
```

```python
rows=[]
for book in books:
    row={}
#     print('---------------')
    row['rank']=book.find(class_='number').text.strip()
    row['title']=book.find('span',itemprop='name').text
    row['author']= book.find(class_='authorName').text.strip()
    row['score'] =book.find(class_='smallText uitext').find('a').text.strip()
    row['votes'] = book.find_all('a')[4].text.strip()
    row['rating'] = book.find(class_='minirating').text.strip()
    rows.append(row)
df=pd.DataFrame(rows)
df
```

```python
df=df[['rank','title','author','score','votes','rating']]
df
```

```python
df.to_csv('female_sci_fi.csv',index=False)
```

```python
df['score'] = df.score.str.replace("score:", "")
```

```python
df['Title'] = df.title.str.extract(r"([^\(]+)")
```

```python
df['Series'] = df.title.str.extract("\((.*),")
df
```

```python
df['Number in Series'] = df.title.str.extract("\(.*,\s[#](.)\)")
```

```python
df['avg_rating']=df.rating.str.extract(r"(.*) avg rating")
df['number_of_ratings']=df.rating.str.extract(r".* — (.*) ratings")
df=df.drop(columns=['title','rating'])
df
```
