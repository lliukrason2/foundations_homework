# Billboard Hot 100
looking for rank, artist, song name

```python
import pandas as pd
import requests
import re
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

%matplotlib inline
```

```python
url='https://www.billboard.com/charts/hot-100'
```

```python
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
charts = doc.find_all(class_='chart-list-item')
```

```python
rows=[]
for chart in charts:
    row={}
#     print('----------')
    row['name'] = chart.find(class_='chart-list-item__title-text').text.strip()
    row['artist'] = chart.find(class_='chart-list-item__artist').text.strip()
    row['rank'] = chart.find(class_='chart-list-item__rank').text.strip()
    rows.append(row)

df=pd.DataFrame(rows)
df
```

```python
df.to_csv('billboard_100.csv',index=False)
```
